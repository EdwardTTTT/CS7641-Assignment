```python
weights_5 = [10, 5, 2, 8, 3]
values_5 = [1, 2, 3, 4, 5]
state_5 = np.array([0, 1, 0, 1, 0])
fitness_cust_5 = mlrose.Knapsack(weights_5, values_5, max_weight_pct_5)


#Knapsack problem (length 5)
problem_KS_L5= mlrose.DiscreteOpt(length = 5, fitness_fn = fitness_cust_5, maximize = True)

# Solve problem using simulated annealing+ Randome Hill Climb + Genetic Algorithm + MIMIC
# Define decay schedule
schedule = mlrose.ExpDecay()
start_time = time.time()
best_state_sa_KS_L5, best_fitness_sa_KS_L5,fitness_curve_sa_KS_L5= mlrose.simulated_annealing(problem_KS_L5, schedule = schedule,
                                                      max_attempts = 500, max_iters = 500,init_state=state_5,
                                                      random_state = 42,curve = True)
stop_time = time.time()
running_time_sa_KS_L5=stop_time- start_time

print('running_time_sa_KS_L5:'+ f"{running_time_sa_KS_L5:.4f}s")


start_time = time.time()
best_state_RHC_KS_L5, best_fitness_RHC_KS_L5,fitness_curve_RHC_KS_L5= mlrose.random_hill_climb(problem_KS_L5, init_state=state_5,
                                                      max_attempts = 500, max_iters = 500,random_state = 42,curve = True)
stop_time = time.time()
running_time_RHC_KS_L5=stop_time- start_time

print('running_time_RHC_KS_L5:'+ f"{running_time_RHC_KS_L5:.4f}s")

start_time = time.time()
best_state_GA_KS_L5, best_fitness_GA_KS_L5,fitness_curve_GA_KS_L5= mlrose.genetic_alg(problem_KS_L5, mutation_prob = 0.1,
                                                      max_attempts = 500, max_iters = 500,random_state = 42,curve = True)
stop_time = time.time()
running_time_GA_KS_L5=stop_time- start_time

print('running_time_GA_KS_L5:'+ f"{running_time_GA_KS_L5:.4f}s")

start_time = time.time()
best_state_mimic_KS_L5, best_fitness_mimic_KS_L5,fitness_curve_mimic_KS_L5= mlrose.mimic(problem_KS_L5, 
                                                      max_attempts = 500, max_iters = 500, random_state = 42,curve = True)
stop_time = time.time()
running_time_mimic_KS_L5=stop_time- start_time

print('running_time_mimic_KS_L5:'+ f"{running_time_mimic_KS_L5:.4f}s")
plt.plot(fitness_curve_sa_KS_L5, '.-', color='red',label='Simulated Annealing')
plt.plot(fitness_curve_RHC_KS_L5, '.-', color='blue',label='Random Hill Climbing')
plt.plot(fitness_curve_GA_KS_L5, '.-', color='green',label='Generic Algorithm')
plt.plot(fitness_curve_mimic_KS_L5, '.-', color='yellow',label='Mimic')

plt.title("Comparison of Fitness Curves for the Knapsack(length 5)")
plt.xlabel('Number of Iterations')
plt.ylabel('Fitness Score')
plt.legend(loc='best')

plt.show()
```
