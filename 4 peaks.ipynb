```python
import six
import sys
sys.modules['sklearn.externals.six'] = six
import  mlrose 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import random
```

```python
fitness_cust = mlrose.FourPeaks(t_pct=0.15)
```

```python
# Four peaks problem (length 10)
problem_4P_L10 = mlrose.DiscreteOpt(length = 10, fitness_fn = fitness_cust, maximize = True, max_val = 10)

# Solve problem using simulated annealing+ Randome Hill Climb + Genetic Algorithm + MIMIC
# Define decay schedule
schedule = mlrose.ExpDecay()
start_time = time.time()
best_state_sa_4P_L10, best_fitness_sa_4P_L10,fitness_curve_sa_4P_L10 = mlrose.simulated_annealing(problem_4P_L10, schedule = schedule,
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_sa_4P_L10=stop_time- start_time

print('running_time_sa_4P_L10:'+ f"{running_time_sa_4P_L10:.4f}s")


start_time = time.time()
best_state_RHC_4P_L10, best_fitness_RHC_4P_L10,fitness_curve_RHC_4P_L10 = mlrose.random_hill_climb(problem_4P_L10,
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_RHC_4P_L10=stop_time- start_time

print('running_time_RHC_4P_L10:'+ f"{running_time_RHC_4P_L10:.4f}s")

start_time = time.time()
best_state_GA_4P_L10, best_fitness_GA_4P_L10,fitness_curve_GA_4P_L10 = mlrose.genetic_alg(problem_4P_L10, mutation_prob = 0.1,
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_GA_4P_L10=stop_time- start_time

print('running_time_GA_4P_L10:'+ f"{running_time_GA_4P_L10:.4f}s")

start_time = time.time()
best_state_mimic_4P_L10, best_fitness_mimic_4P_L10,fitness_curve_mimic_4P_L10 = mlrose.mimic(problem_4P_L10, 
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_mimic_4P_L10=stop_time- start_time

print('running_time_mimic_4P_L10:'+ f"{running_time_mimic_4P_L10:.4f}s")

plt.plot(fitness_curve_sa_4P_L10, '.-', color='red',label='Simulated Annealing')
plt.plot(fitness_curve_RHC_4P_L10, '.-', color='blue',label='Random Hill Climbing')
plt.plot(fitness_curve_GA_4P_L10, '.-', color='green',label='Generic Algorithm')
plt.plot(fitness_curve_mimic_4P_L10, '.-', color='yellow',label='Mimic')

plt.title("Comparison of Fitness Curves for the 4 peaks(length 10)")
plt.xlabel('Number of Iterations')
plt.ylabel('Fitness Score')
plt.legend(loc='best')

plt.show()
```
