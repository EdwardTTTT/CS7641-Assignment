```python
import six
import sys
sys.modules['sklearn.externals.six'] = six
import  mlrose 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import random
```

```python
# Define alternative N-Queens fitness function for maximization problem
def queens_max(state):

    # Initialize counter
    fitness_cnt = 0

    # For all pairs of queens
    for i in range(len(state) - 1):
         for j in range(i + 1, len(state)):
                # Check for horizontal, diagonal-up and diagonal-down attacks
                if (state[j] != state[i]) \
                and (state[j] != state[i] + (j - i)) \
                and (state[j] != state[i] - (j - i)):
                    # If no attacks, then increment counter
                    fitness_cnt += 1
                
    return fitness_cnt
# Initialize custom fitness function object
fitness_cust = mlrose.CustomFitness(queens_max)
```

```python
#8-queen problem (length 8)
problem_8Q = mlrose.DiscreteOpt(length = 8, fitness_fn = fitness_cust, maximize = True, max_val = 8)

# Solve problem using simulated annealing+ Randome Hill Climb + Genetic Algorithm + MIMIC
# Define decay schedule
schedule = mlrose.ExpDecay()
start_time = time.time()
best_state_sa_8Q, best_fitness_sa_8Q,fitness_curve_sa_8Q = mlrose.simulated_annealing(problem_8Q, schedule = schedule,
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_sa_8Q=stop_time- start_time

print('running_time_sa_8Q:'+ f"{running_time_sa_8Q:.4f}s")


start_time = time.time()
best_state_RHC_8Q, best_fitness_RHC_8Q,fitness_curve_RHC_8Q = mlrose.random_hill_climb(problem_8Q,
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_RHC_8Q=stop_time- start_time

print('running_time_RHC_8Q:'+ f"{running_time_RHC_8Q:.4f}s")

start_time = time.time()
best_state_GA_8Q, best_fitness_GA_8Q,fitness_curve_GA_8Q = mlrose.genetic_alg(problem_8Q, mutation_prob = 0.1,
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_GA_8Q=stop_time- start_time

print('running_time_GA_8Q:'+ f"{running_time_GA_8Q:.4f}s")

start_time = time.time()
best_state_mimic_8Q, best_fitness_mimic_8Q,fitness_curve_mimic_8Q = mlrose.mimic(problem_8Q, 
                                                      max_attempts = 500, max_iters = 1000,random_state = 42,curve = True)
stop_time = time.time()
running_time_mimic_8Q=stop_time- start_time

print('running_time_mimic_8Q:'+ f"{running_time_mimic_8Q:.4f}s")

plt.plot(fitness_curve_sa_8Q, '.-', color='red',label='Simulated Annealing')
plt.plot(fitness_curve_RHC_8Q, '.-', color='blue',label='Random Hill Climbing')
plt.plot(fitness_curve_GA_8Q, '.-', color='green',label='Generic Algorithm')
plt.plot(fitness_curve_mimic_8Q, '.-', color='yellow',label='Mimic')

plt.title("Comparison of Fitness Curves for the 8-Queen(length 8)")
plt.xlabel('Number of Iterations')
plt.ylabel('Fitness Score')
plt.legend(loc='best')

plt.show()
```
